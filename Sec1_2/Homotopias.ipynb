En clase se propusieron las siguientes curvas paramétricas:

\begin{align}
f_0 &= (cosx , senx)\\
f_1 &= (cos(2\pi - x) , sen(2\pi - x))\\
\end{align}

las cuales parametrizan media circunferencia unitaria.

Se propuso la siguiente función como homotopía:

\begin{equation}
h_t(x) = H(x,t) = (1 - t)f_0 + tf_1
\end{equation}

la cual se puede escribir en forma paramétrica de la siguiente manera:

\begin{align}
h_t(x) &= (1 - t)f_0 + tf_1 \\
&= ((1 - t)\, cosx , (1 - t)\, senx) + ( t\, cos(2\pi - x), t \, sen(2\pi - x) \\
&= ((1-t)cosx + tcos(2\pi - x) , (1-t)senx + tsen(2\pi - x))
\end{align}


Y graficamos para valores de $t$ en $[0,1]$

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import *
%matplotlib widget
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(0, np.pi, 100)
for t in np.linspace(0,1,20):

    x1 = (1-t)*np.cos(x) + t*np.cos(2*np.pi - x)
    y1 = (1-t)*np.sin(x) + t*np.sin(2*np.pi - x)

    plt.plot(x1, y1,  label = f'line {round(t,1)}')
plt.legend()
ax.set_ylim(-1.01,1.01)
ax.set_xlim(-1.01,1.01)
plt.show()
```

También se propuso la siguiente función, la cual se puede escribir en forma paramétrica de la siguiente manera:

\begin{align}
h_t(x) = ( cos(2t\pi - (1-2t)x), \, sen(2t\pi - (1-2t)x) ) 
\end{align}


Y graficamos para valores de $t$ en $[0,1]$:

Nota. La siguente es una gráfica es interactiva, debe aparecer un slider que permita mover el valor del parámetro $t$.

```python
x = np.linspace(0, np.pi, 100)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x1 = np.cos(x)
y1 = np.sin(x)
line, = ax.plot(x1, y1)
ax.set_ylim(-1.01,1.01)
ax.set_xlim(-1.01,1.01)
def update(t = 0.0):
    x1 = np.cos(2*t*np.pi + (1 - 2*t)*x)
    y1 = np.sin(2*t*np.pi + (1 - 2*t)*x)
    line.set_data(x1, y1)
    fig.canvas.draw_idle()

interact(update);
```

Vemos que la segunda familia de curvas paramétricas forman arcos de la circunferencia unitaria que van cambiando de longitud de arco, en $t=0$ se tiene el arco representado por $f_0$, luego tendiendo a cero hasta anularse en $t = 1/2$ y luego vuelve a crecer hasta igualar al arco representado por $f_1$ en $t=1$.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(0, np.pi, 100)
for t in np.linspace(0,1,20):

    x1 = (1/(1+t))*np.cos(x) + t*np.cos(2*np.pi - x)
    y1 = (1/(1+t))*np.sin(x) + t*np.sin(2*np.pi - x)

    plt.plot(x1, y1,  label = f'line {round(t,1)}')
plt.legend()
ax.set_ylim(-2.01,2.01)
ax.set_xlim(-2.01,2.01)
plt.show()
```
